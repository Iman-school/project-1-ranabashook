```python
x = Insurance_data["age"]
y = Insurance_data["bmi"]

plt.scatter(x, y)
plt.xlabel("age")
plt.ylabel("bmi")
```

```python
from sklearn.linear_model import LinearRegression 

x = Insurance_data["age"].values.reshape(-1,1)
y = Insurance_data["bmi"]

lr_model = LinearRegression()
lr_model.fit(x, y)
y_pred = lr_model.predict(x)
y_pred 

plt.scatter(x, y)
plt.xlabel("age")
plt.ylabel("bmi")

plt.plot(x, y_pred)
```

```python
theta_0 = lr_model.intercept_
theta_1 = lr_model.coef_
theta_0, theta_1
```
